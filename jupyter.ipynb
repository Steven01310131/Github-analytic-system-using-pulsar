```python
import json 
import requests
from pprint import pprint
```

```python
url=f"https://api.github.com/search/repositories?q=created:>2022-05-01+archived:false"
user_data = requests.get(url).json()
```

```python
pprint(user_data)
```

```python
for i in range(30):
    repo_name=user_data['items'][i]['full_name']
    print(repo_name)
```

```python

spec_repo_url=requests.get(f"https://api.github.com/repos/{repo_name}/commits?per_page=1&page=1")
headers = spec_repo_url.headers
```

```python
pprint(int(headers['Link'].split('&page=')[1].split(';')[0].split('>')[0]))
```
